```python
# @title
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
```

# Display the Five Rows

```python
print("First five rows:")
print(iris_df.head())
```

# Display the shape of the dataset

```python
print("This is the My Shape DataSet:")
print(iris_df.shape)
```

# Display summary statistics


```python
print("This is the Summerized Way of statistics ")
print(iris_df.describe())
```

# Here I am Using the Style of Data Visualization


```python
sns.set(style='whitegrid')
```

# Pair of Plot

```python
plt.figure(figsize=(8, 4))
plt.title('This is the Pairing and ploting the Iris Dataset')
sns.pairplot(iris_df, hue='species', palette='Set1', markers=['o', 's', 'D'])

plt.show()
```

# This is My Plot boxes

```python
plt.figure(figsize=(4, 3))
sns.boxplot(x='species', y='sepal length (cm)', data=iris_df)
plt.title('Box plot of Sepal lenght and Species')
plt.xlabel('Species')
plt.ylabel('Sepal Lenght')
plt.show()
```

# Violin Ploting Here i am Using

```python
plt.figure(figsize=(8, 3))
sns.violinplot(x='species', y='petal length (cm)', data=iris_df, inner='quartile', palette='Set1')
plt.title('Here i am Using Viloin Ploting')
plt.show()
```

# Using Scatter Chart for Visualization

```python
plt.figure(figsize=(8, 3))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=iris_df, style='species', palette='Set1')
plt.title('Ploting of Sepal Lenght and Species')
plt.xlabel('Sepal Lenght')
plt.ylabel('Sepal Width')
plt.show()
```

#Here i am Using Histogram for Simple to Understanding

```python
plt.figure(figsize=(5, 4))
sns.histplot(data=iris_df, x='sepal length (cm)', hue='species', multiple='stack', bins=10, kde=True)
plt.title('Using Histogram ')
plt.xlabel('Sepal Length ')
plt.ylabel('Frequency')
plt.show()
```

# Count Ploting for Species


```python
plt.figure(figsize=(4, 3))
sns.countplot(x='species', data=iris_df, palette='Set1')
plt.title('Count every Total Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1, 2], labels=iris.target_names)
plt.show()
```

# Heatmap of Cor-relation of met

```python
plt.figure(figsize=(6, 3))
correlation_matrix = iris_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Using Heatmap for Co-relation')
plt.show()
```
